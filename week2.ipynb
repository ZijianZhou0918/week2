# 3. Hypothesis and Investigate

### Hypothesis 1  - Are customers mainly concentrated in a few specific age groups?


### Investigate Method:

- Divide users into age groups, with each group spanning 10 years, and observe the distribution of the data.

```python
df = df_cus.copy()
age_groups = pd.cut(df['Age'], bins=range(0, 100, 10), right=False)
age_counts = age_groups.value_counts().sort_index()

# plot - sns bar plot
sns.set_palette("Blues_r")
sns.barplot(x=age_counts.index.astype(str), y=age_counts.values)
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.title('Overall Customer Age Distribution')
plt.show()
```

### Findings:
- From this analysis, we can conclude that our target market consists primarily of customers in the age groups of `20-30` and `30-40`. These age groups typically comprise young adults and young families, making them potential target markets. Given their age range, these segments tend to have higher transportation needs and may rely more frequently on taxi services.

- Customers in the 20-30 and 30-40 age groups share certain characteristics and preferences. For example, they may prioritize convenience, affordability, mobile payment options, and comfortable travel experiences. Understanding these user traits can help determine the key areas of focus in terms of service and operations.

### Hypothesis 2  - Are the user groups of these two companies different?

- Consider `Gender`, `Age` and `Income`

### Step 1 : Get the merged table - combine all sub dataset

```python
merged_df = pd.merge(df_cus, df_trans, on='Customer ID')
merged_df = pd.merge(merged_df, df_cab, on='Transaction ID')
merged_df = pd.merge(merged_df, df_city, on='City')
merged_df.head()
```

### Step 2: Segmenting user groups based on cab companies
Please note that the same user may take multiple rides, and a single user may take rides with multiple companies. 

Therefore, we need to consider each user in different companies only once.

```python
df = merged_df.copy()
df_p = df[df['Company']=='Pink Cab'].drop_duplicates(subset='Customer ID', keep='first')
df_y = df[df['Company']=='Yellow Cab'].drop_duplicates(subset='Customer ID', keep='first')
```

### Step 3: Are the age groups of users different between the two companies?

```python
age_groups = pd.cut(df_p['Age'], bins=range(10, 80, 10), right=False)
age_counts = age_groups.value_counts().sort_index()

# plot - sns bar plot
age_counts.plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.title('Age Distribution - Pink Cab')

plt.figure(figsize=(8,4))
sns.set_palette("Blues_r")

sns.barplot(x=age_counts.index.astype(str), y=age_counts.values)
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.title('Customer Age Distribution - Pink Cab')
plt.show()


age_groups = pd.cut(df_y['Age'], bins=range(10, 80, 10), right=False)
age_counts = age_groups.value_counts().sort_index()

# plot - sns bar plot

age_counts.plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.title('Age Distribution - Yellow Cab')
sns.set_palette("Blues_r")
plt.figure(figsize=(8,4))
sns.barplot(x=age_counts.index.astype(str), y=age_counts.values)
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.title('Customer Age Distribution - Yellow Cab')
plt.show()
```

- The age distribution of the user groups of the two taxi companies is similar, but Yellow Cab has a larger presence than Pink Cab in every age group. This suggests that Yellow Cab holds a market share advantage and has a broader user base. By increasing investment in Yellow Cab, we can further expand your market share and acquire more users.

### Step 4: Are the Income of users different between the two companies?

```python
income_groups = pd.cut(df_p['Income (USD/Month)'], bins=range(0, 40000, 5000), right=False)
income_counts = income_groups.value_counts().sort_index()

# plot - sns bar plot
income_counts.plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.title('Income Distribution - Pink Cab')

plt.figure(figsize=(15,4))
sns.set_palette("Blues_r")

sns.barplot(x=income_counts.index.astype(str), y=income_counts.values)
plt.xlabel('income Groups')
plt.ylabel('Count')
plt.title('Customer income Distribution - Pink Cab')
plt.show()


income_groups = pd.cut(df_y['Income (USD/Month)'], bins=range(0, 40000, 5000), right=False)
income_counts = income_groups.value_counts().sort_index()

# plot - sns bar plot
income_counts.plot(kind='pie', autopct='%1.1f%%')
plt.axis('equal')
plt.title('Income Distribution - Yellow Cab')

plt.figure(figsize=(15,4))
sns.set_palette("Blues_r")

sns.barplot(x=income_counts.index.astype(str), y=income_counts.values)
plt.xlabel('income Groups')
plt.ylabel('Count')
plt.title('Customer income Distribution - Yellow Cab')
plt.show()
```

- The user groups' income distribution of the two taxi companies is similar, but Yellow Cab has a higher income among the main income segments compared to Pink Cab. This suggests that Yellow Cab has an advantage in the high-income demographic.

### Hypothesis 3: Are there differences in the consumption habits of taxi customers in different cities？
- Group the data by city and calculate the average spending level.

```python
import pandas as pd
import matplotlib.pyplot as plt

average_spending = merged_df.groupby('City')['Price Charged'].mean()
for city, spending in average_spending.items():
    print(f"City: {city}：Average Spending: {spending}")
plt.figure(figsize=(15, 6))
average_spending.plot(marker='o', linestyle='-', color='b')

plt.title('Average Spending by City')
plt.xlabel('City')
plt.ylabel('Average Spending')
plt.xticks(range(len(average_spending)), average_spending.index, rotation=45)
for i, spending in enumerate(average_spending):
    plt.annotate(f'{spending:.2f}', (i, spending+1), textcoords="offset points", xytext=(0,5), ha='center')
    
plt.show()
```

### Findings
Based on the provided average spending data, we can draw some preliminary conclusions:

1. There are variations in the average spending levels across different cities. For example, New York NY has a higher average spending of 570.20, while Nashville TN has a lower average spending of 318.64.


2. Some cities exhibit higher average spending, such as Dallas TX and Silicon Valley, with average spendings of 447.83 and 432.46, respectively.


3. Certain cities have lower average spending, like Pittsburgh PA and Sacramento CA, with average spendings of 337.04 and 321.76, respectively.


4. Cities with higher average spending may not necessarily be the best investment choice, as other factors such as market potential and competitive environment need to be considered.

In conclusion, the average spending data provides insights into the consumer habits across different cities to consider. However, further analysis of other datasets and consideration of additional key factors are necessary before making a final investment decision.

### Hypothesis 4：Does a larger city population mean more users?
- Sort the cities based on population and plot a scatter plot between population and number of customers.

```python
import pandas as pd
import matplotlib.pyplot as plt

df_city_tmp = df_city.copy()

df_city_tmp['Population'] = df_city_tmp['Population'].apply(lambda x:int(x.replace(',','')))
df_city_tmp['Users'] = df_city_tmp['Users'].apply(lambda x:int(x.replace(',','')))

df_city_tmp = df_city_tmp.sort_values(by='Population')

customer_count = df_city_tmp['Users']
population = df_city_tmp['Population']
plt.scatter(population, customer_count)

plt.title('Relationship between Population and Customer Count')
plt.xlabel('Population')
plt.ylabel('Customer Count')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

### Findings
1. There is a positive correlation between population and users: The overall trend shows that cities with larger populations tend to have more users. This suggests that cities with larger populations provide a larger user base and potential market.


2. Some cities have a higher number of users than expected: In cities with smaller populations, there are a few cities that have relatively high numbers of users, surpassing other cities of similar size. For example, cities like "NEW YORK NY," "SAN FRANCISCO CA," and "BOSTON MA" have relatively high numbers of users, potentially due to their significant roles in technology, innovation, and culture.


3. Some cities have relatively lower numbers of users: Despite having larger populations, there are several cities with relatively lower numbers of users. For instance, cities like "MIAMI FL," "SILICON VALLEY," and "ORANGE COUNTY" have smaller user populations despite their larger populations. This could be attributed to other factors such as economic conditions, internet penetration rates, and more.

### Hypothesis 5：Is there any seasonality in number of customers using the cab service?
- acf and pacf plot

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import warnings
warnings.filterwarnings("ignore")

date_data = merged_df.groupby('Date of Travel').count()['Customer ID']
date_data_cvt = date_data.reset_index()
date_data_cvt.columns = ['Date','Count']

plt.figure(figsize=(15,8))
plt.scatter(date_data_cvt['Date'],date_data_cvt['Count'],s=5)
plt.title('Date and count for cab using')
plt.show()

# 绘制ACF图
plt.figure(figsize=(10, 5))
plot_acf(date_data, lags=20)  
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')

plt.show()
```

### Findings

We can observe that the usage of taxis exhibits clear seasonality because there are distinct periodic peaks in the ACF (Autocorrelation Function) plot. 

The significant values of the autocorrelation coefficients at specific lags indicate the presence of repetitive patterns in the data across different time intervals. The occurrence of these periodic peaks suggests fluctuations in the demand for taxi services between different seasons, which may be influenced by factors such as seasonal changes, holidays, weather conditions, and other relevant factors.
